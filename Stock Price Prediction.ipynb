    def process_data(data , n_features):
        dataX, dataY = [], [] 
        for i in range(len(data)-n_features):
            a = data[i:(i+n_features), 0]
            dataX.append(a) 
            dataY.append(data[i + n_features, 0])
        return np.array(dataX), np.array(dataY)
